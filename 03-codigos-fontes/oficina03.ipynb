## Exercício 2

Considere a função definida abaixo:

$$
f(x) = x \cdot \sin(10 \pi x) + 1
$$

Seu objetivo é **encontrar o valor de $x \in [-1, 2]$ que **maximiza** essa função.

#### Gráfico da Função

A função possui múltiplos máximos locais, o que torna o problema adequado para métodos que evitam ficar presos em ótimos locais. O gráfico mostra uma oscilação periódica com amplitude variável dentro do intervalo dado.

#### Tarefa

Implemente algoritmos de otimização local para encontrar um valor aproximado de $x$ que maximize $f(x)$. Compare o desempenho entre:

- Simulated Annealing
- Busca Tabu

#### Requisitos

1. Registre e compare:
   - Melhor valor de $x$ encontrado
   - Valor de $f(x)$ correspondente
   - Número de iterações até convergência (se aplicável)


```python
import numpy as np
import matplotlib.pyplot as plt

# Função objetivo: f(x) = x * sin(10 * pi * x) + 1, definida em x ∈ [-1, 2]
def funcao_objetivo(x):
    return x * np.sin(10 * np.pi * x) + 1

# Domínio da função
dominio = (-1, 2)

# Plot da função
x_vals = np.linspace(dominio[0], dominio[1], 1000)
y_vals = funcao_objetivo(x_vals)

plt.plot(x_vals, y_vals)
plt.title("Função Objetivo: f(x) = x * sin(10πx) + 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
